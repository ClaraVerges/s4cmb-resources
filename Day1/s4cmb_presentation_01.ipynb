Presentation of the library
====
Welcome! s4cmb is a package to study instrumental systematic effects in the context of current and future Cosmic Microwave Background experiments. This series of notebooks describes the main parts of the s4cmb library.
More specifically it focuses on the following:
* Instrument
* Input sky
* Scanning strategy
* Time-Ordered data
* Output sky

A typical pipeline can be seen as following:

```python
from IPython.display import Image
Image(filename='../data/presentation.png')
```

First you define the instrument, and its properties (pointing, beams, and so on). Then you pick up a scanning strategy, that is you define where and when to look on the sky. From all three, you scan a simulated sky to produce Time-Ordered Data (or timestreams), and then you project back on the sky to make maps of the observations.
The injection of instrumental systematic effects can be done at all those steps!
